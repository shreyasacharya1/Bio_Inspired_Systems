```python
import random
import math

# Objective function
def f(x):
    return x * math.sin(10 * math.pi * x) + 1

# Get neighbors in a grid (Moore neighborhood)
def get_neighbors(i, j, grid_size):
    neighbors = []
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            ni, nj = i + dx, j + dy
            if 0 <= ni < grid_size and 0 <= nj < grid_size:
                neighbors.append((ni, nj))
    return neighbors

def parallel_cellular_algorithm():
    grid_size = 5
    iterations = 30

    # Initialize grid with random solutions
    grid = [[random.random() for _ in range(grid_size)] for _ in range(grid_size)]

    for _ in range(iterations):
        new_grid = [[0 for _ in range(grid_size)] for _ in range(grid_size)]

        for i in range(grid_size):
            for j in range(grid_size):
                cell = grid[i][j]
                best = cell
                best_f = f(cell)

                # Find best neighbor
                for ni, nj in get_neighbors(i, j, grid_size):
                    val = grid[ni][nj]
                    if f(val) > best_f:
                        best = val
                        best_f = f(val)

                # Update by moving toward best neighbor
                new_grid[i][j] = (cell + best) / 2

        grid = new_grid  # update all cells

    # Find global best
    best_val = 0
    best_x = 0
    for i in range(grid_size):
        for j in range(grid_size):
            if f(grid[i][j]) > best_val:
                best_val = f(grid[i][j])
                best_x = grid[i][j]

    print("Best x:", best_x)
    print("Best fitness:", best_val)

parallel_cellular_algorithm()
```
