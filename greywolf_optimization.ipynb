```python
import random
import math

# Objective function
def f(x):
    return x * math.sin(10 * math.pi * x) + 1

# Grey Wolf Optimizer
def gwo():
    wolves = 20
    iterations = 50

    # Initialize population
    population = [random.random() for _ in range(wolves)]
    fitness = [f(x) for x in population]

    # Initialize alpha, beta, delta
    alpha = beta = delta = 0
    alpha_f = beta_f = delta_f = -1

    # Find initial leaders
    for i in range(wolves):
        if fitness[i] > alpha_f:
            delta_f = beta_f
            delta = beta
            beta_f = alpha_f
            beta = alpha
            alpha_f = fitness[i]
            alpha = population[i]
        elif fitness[i] > beta_f:
            delta_f = beta_f
            delta = beta
            beta_f = fitness[i]
            beta = population[i]
        elif fitness[i] > delta_f:
            delta_f = fitness[i]
            delta = population[i]

    # Main loop
    for t in range(iterations):
        a = 2 - 2 * (t / iterations)  # decreases linearly 2 â†’ 0

        for i in range(wolves):
            x = population[i]

            # Update w.r.t alpha
            A1 = 2 * a * random.random() - a
            C1 = 2 * random.random()
            D1 = abs(C1 * alpha - x)
            X1 = alpha - A1 * D1

            # Update w.r.t beta
            A2 = 2 * a * random.random() - a
            C2 = 2 * random.random()
            D2 = abs(C2 * beta - x)
            X2 = beta - A2 * D2

            # Update w.r.t delta
            A3 = 2 * a * random.random() - a
            C3 = 2 * random.random()
            D3 = abs(C3 * delta - x)
            X3 = delta - A3 * D3

            # New position
            new_x = (X1 + X2 + X3) / 3
            new_x = max(0, min(1, new_x))  # bound protection

            population[i] = new_x
            fitness[i] = f(new_x)

        # Update leaders
        for i in range(wolves):
            if fitness[i] > alpha_f:
                delta_f = beta_f
                delta = beta
                beta_f = alpha_f
                beta = alpha
                alpha_f = fitness[i]
                alpha = population[i]
            elif fitness[i] > beta_f:
                delta_f = beta_f
                delta = beta
                beta_f = fitness[i]
                beta = population[i]
            elif fitness[i] > delta_f:
                delta_f = fitness[i]
                delta = population[i]

    print("Best wolf (alpha):", alpha)
    print("Best fitness:", alpha_f)

gwo()
```
